```python
#removing columns not useful in project
#useful columns: movie_id,title,keyword,genre,overview,cast,crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
#to check for duplicate data
movies.duplicated().sum()
```

```python
movies=movies.drop_duplicates()
```

```python
movies.duplicated().sum()
#duplicate data removed
```

```python
#to convert string to integer format
import ast
```

```python
#creating list of genres for every movie
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#same for cast but for cast we only require three main casts
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
#show three main cast members only
```

```python
# from crew we only require director
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#to replace any sapce between any name
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
#to create tags by merging all other columns
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
#converting list to string
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'][0]
```

```python
#to convert tags to vectors(vectors will be formed using most frequent words)
#stop_words removes words in english that have no meaning
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new['tags']).toarray()
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
similarity = cosine_similarity(vector)
```

```python
#tells similarity of first movie with all other movies
similarity[0]
```

```python
#when a movie is given func returns 5 other recomended movies
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
